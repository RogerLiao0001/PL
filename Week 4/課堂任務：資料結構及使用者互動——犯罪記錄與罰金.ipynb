# 題目：  
已知有甲、乙、丙三人，分別分別年齡為40、52、63，他們分別觸犯多條道路交通管理處罰條例。  
<br>
甲：第62條 汽車駕駛人駕駛汽車肇事，被罰了2000元、 第12條 汽車使用吊銷、註銷之牌照，被罰了5000元  
乙：第16條 裝置高音量喇叭或其他產生噪音器物，被罰了1200元、 第56條 在禁止臨時停車處所停車，被罰了800元、 第82條 未經許可在道路擺設攤位，被罰了1500元、 第17條 汽車不依限期參加定期檢驗或臨時檢驗，被罰了900元  
丙：第21條 未領有駕駛執照駕駛小型車或機車，被罰了8000元  
<br>
請使用while迴圈讓使用者選擇： 1.查看甲資料 2.查看乙資料 3.查看丙資料 4.犯罪數排名 5.累積罰金&罰鍰數排名 6.結束程式  
其餘(包含文字)顯示請重新輸入

```python
#建立索引
甲 ={
'姓名':"甲",
'年齡':40,
'第62條 汽車駕駛人駕駛汽車肇事':2000,
'第12條 汽車使用吊銷、註銷之牌照':5000
}
乙 ={
'姓名':"乙",
'年齡':52,
'第16條 裝置高音量喇叭或其他產生噪音器物':1200,
'第56條 在禁止臨時停車處所停車':800,
'第82條 未經許可在道路擺設攤':1500,
'第17條 汽車不依限期參加定期檢驗或臨時檢驗':900
}
丙 ={
'姓名':"丙",
'年齡':63,
'第21條 未領有駕駛執照駕駛小型車或機車':8000
}
for i in (甲,乙,丙):
    for k in i:
        if k == "年齡":
            print(k,i[k])
        elif k == "姓名":
            print(i[k],"的資料為")
        else:
            print("曾犯",k,"被罰",i[k],"元")
    print()
            
        
```

```python
while True:
    m = input("請選擇服務項目：1.查看甲資料 2.查看乙資料 3.查看丙資料 4.犯罪數排名 5.累積罰金&罰鍰數排名 6.結束程式")
    if m == "1":
        print(甲)
        print()
    elif m== "2":
        print(乙)
        print()
    elif m=="3":
        print(丙)
        print()
    elif m=="4":
        sorter = [("甲，共",len(甲)-2,"次"), ("乙，共",len(乙)-2,"次"), ("丙，共",len(丙)-2,"次") ]
        print("少到多：",sorted(sorter,key=lambda x: x[1]))
        print()
    elif m=="5":
        a=int (0)
        b=int (0)
        c=int (0)
        for i in (甲,乙,丙):
            for k in i:
                if (k != "年齡")and(k != "姓名"):
                    if i == 甲:
                        a=a+int (i[k])
                    elif i == 乙:
                        b=b+int (i[k])
                    elif i == 丙:
                        c=c+int (i[k])
        sorter2 = [("甲，共",a,"元"), ("乙，共",b,"元"), ("丙，共",c,"元") ]
        print("少到多：",sorted(sorter2,key=lambda x: x[1]))
        print()
    elif m=="6":
        print("感謝使用")
        break
    else:
        print("輸入錯誤，請重新輸入")
        print()
```
