```python
import jieba
import jieba.analyse
import pandas as pd
import numpy
from collections import Counter
```

```python
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
data = pd.read_csv('不會被取代.csv')
data
```

```python
allwords = []
for content in data:
    allwords.append(content)
allwords
```

```python
alljie = [] 
for i in allwords:
    seg = jieba.analyse.extract_tags(i, 50)
    alljie.append(seg)
alljie
```

```python
flat = [x for sublist in alljie for x in sublist]
diction = Counter(flat)
font = "Arial Unicode"
mask = numpy.array(Image.open("公事包7.png"))
wordcloud = WordCloud(background_color = "white", font_path=font, mask = mask)#, mask = mask)
wordcloud.generate_from_frequencies(diction)
plt.figure(figsize=(400,400))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
