# 作業一：使用Pandas進行資料比較：找出未額滿的通識全英通識
### ・目標1: 快速找出師大全英文、未額滿的通識課程
### ・目標2: 交集：找出未額滿且為全英文的通識課  
### ・目標3: 找出差集、對稱差集、聯集並給予意義 
資料來源：師大課程系統（111學年第1學期資料）

```python
#匯入師大111-1通識資料
import warnings #移除警告，使畫面簡潔
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
file = pd.read_csv("通識表.csv")
file.tail()
```

```python
#前置作業:移出無關資訊
#移除『中文課程名稱、英文課程名稱』括號內不必要字元
file['中文課程名稱'] = file['中文課程名稱'].str.replace(r"\[.*\]","", case = False)
file['英文課程名稱'] = file['英文課程名稱'].str.replace(r"\[.*\]","", case = False)
#移除無意義欄位
file = file.drop(["MOOCS","班","年","限性別","必/選", "全/半"], axis=1, inplace=False,  errors='ignore')  
file.tail()
```

```python
#列出所有全英文課程
En = file.loc[file['全英語'] == '是'] 
print("全英文課程列表：")
En
```

```python
#列出所有未額滿課程(由限修人數、選修人數)
notFull = file.loc[file['限修人數'] > file['選修人數']]
print("未額滿課程列表：")
notFull
```

```python
#比較全部、全英文、未額滿課程數量！
print("全部通識課程數:", len(file)-1, "全英文課程數:", len(En), "未額滿課程數:", len(notFull))
```

```python
#取交集(全英且未額滿課程)：
inter = pd.merge(En, notFull, how ='inner')
print("全英且未額滿的課程有", len(inter),"個!\n")
print("全英且未額滿課程列表：")
inter
```

```python
#取左差集(已額滿的全英課程)：
enDiff = En[En.開課代碼.isin(notFull.開課代碼) == False]
print("已額滿的全英課程有", len(enDiff),"個!\n")
print("已額滿的全英課程列表：")
enDiff
```

```python
#取右差集(未額滿的非全英課程)：
notFullDiff = notFull[notFull.開課代碼.isin(En.開課代碼) == False]
print("未額滿的非全英課程有", len(notFullDiff),"個!\n")
print("未額滿的非全英課程列表：")
notFullDiff
```

```python
#取聯集(未額滿或全英文)：
uni = pd.concat([En, notFull])
uni2 = uni.drop_duplicates()
print("未額滿或全英文課程有", len(uni2),"個!\n")
print("未額滿或全英文課程列表：")
display(uni2)
```

```python
#取對稱差集(未額滿但非英文 或 額滿且全英文)：
sym = uni[uni.開課代碼.isin(inter.開課代碼) == False]
print("未額滿但非英文 或 額滿且全英文的課程有", len(sym),"個!\n")
print("未額滿但非英文 或 額滿且全英文課程列表：")
sym
```
